```python
import pandas as pd
from tinydb import TinyDB 
import numpy as np
from sklearn.model_selection import train_test_split
```

## Preprocessando dados

```python
db_content = TinyDB("db_240123.json")
df = pd.DataFrame(db_content) # cria dataframe a partir do conteudo do arquivo do banco de dados
```

```python
df #conferindo dataframe
```

```python
#df.drop(columns=df.columns[0], axis=1, inplace=True) # deleta primeira coluna
df.drop(columns=["led_color", "led_dimmer"], inplace=True) # nao se preocupa com cor nesse momento
df
```

```python
import time
#df["led_on"] = df["led_on"].apply(lambda x: np.random.choice([0,1])) # preenche coluna led_on com valores binarios randomicos
#df # to check changes

# Change epoch to localtime, maybe change to labels in the future
df["timestamp"] = df["timestamp"].apply(lambda x: time.strftime('%H', time.localtime(x)))
df["timestamp"] = df["timestamp"].astype('category')
```

## Separando bases de treinamento e teste

```python
train, test = train_test_split(df, test_size=0.3, random_state=42, stratify=df["led_on"])
val, test = train_test_split(test, test_size=0.1/0.3, random_state=42, stratify=test["led_on"])

val.to_csv("val.csv", index=False)
train.to_csv("train.csv", index=False)
test.to_csv("test.csv", index=False)
```
